```python
# prompt: IN the above create df, I want to add a column of the name of sharks that invested in the deal. If there is only one shark that has invested write N/A, else write the names of the sharks not the column names.

shark_cols = ['barbara_inv_amt', 'mark_inv_amt', 'lori_inv_amt', 'robert_inv_amt', 'daymond_inv_amt', 'kevin_inv_amt', 'guest_inv_amt']
shark_names = ['Barbara', 'Mark', 'Lori', 'Robert', 'Daymond', 'Kevin', 'Guest']

def get_shark_names(row):
  investors = []
  for i, col in enumerate(shark_cols):
    if row[col] != 0:
      investors.append(shark_names[i])
  if len(investors) == 0:
    return 'N/A'
  elif len(investors) == 1:
    return 'N/A'
  else:
    return ', '.join(investors)

df['shark_names'] = df.apply(get_shark_names, axis=1)
```
